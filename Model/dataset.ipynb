```python
# Load configuration from file.
import os
import yaml

from easydict import EasyDict

config_path = os.path.join("dataset.yaml")
with open(config_path, "r") as stream:
  config = EasyDict(yaml.load(stream, Loader=yaml.FullLoader))
print(f"Dataset Configuration:\n\n{yaml.dump(dict(config)).strip()}")
```

```python
# Load annotation data:
from datasetinsights.datasets.unity_perception import AnnotationDefinitions

annotation_def = AnnotationDefinitions(data_root=config.dataset_root, version=config.schema_version)
definition_dict = annotation_def.get_definition(def_id=config.annotation_id)
print(yaml.dump(definition_dict).strip())
```

```python
# Load metadata for dataset.
from datasetinsights.datasets.unity_perception import Captures
captures = Captures(data_root=config.dataset_root, version=config.schema_version)
captures_df = captures.filter(def_id=config.annotation_id)
captures_df
```
